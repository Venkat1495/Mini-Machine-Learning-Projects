# Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Importing the Dataset

```python
path = "Salary_Data_slr.csv"
df = pd.read_csv(path)
df.head
```

```python
x = df.iloc[:, 0:1].values
y = df.iloc[:, 1].values
print(x)
print(y)
```

# Splitting Dataset into Traning set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

# Traning the Simple Linear Regression model on the Traning set

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(X_train,Y_train)
```

# Prediciting the Test Set Results

```python
Y_Predict = clf.predict(X_test)
print(Y_Predict)
```

# Visualising the Traning Set Results

```python
plt.scatter(X_train,Y_train, color= 'purple')
plt.plot(X_train, clf.predict(X_train), color='brown')
plt.title('Salary VS Exeprience (Training Set)')
plt.xlabel('Years Of Exeprience')
plt.ylabel('Salary')
plt.show()
```

# Visualising the Test Set Results

```python
plt.scatter(X_test,Y_test, color= 'purple')
plt.plot(X_train, clf.predict(X_train), color='brown')
plt.title('Salary VS Exeprience (Test Set)')
plt.xlabel('Years Of Exeprience')
plt.ylabel('Salary')
plt.show()
```
