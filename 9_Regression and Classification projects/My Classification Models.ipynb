# Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Importing the dataset

```python
path = "Data_classification copy.csv"
df = pd.read_csv(path)
df.head
```

```python
x = df.iloc[:, 0:10].values
y = df.iloc[:, 10].values
print(x)
print(y)
```

# Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=10)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(Y_train)
```

```python
print(Y_test)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

# Decision Tree Classification Model

## Training model on the Training data set

```python
from sklearn.tree import DecisionTreeClassifier
classifier_DTC = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier_DTC.fit(X_train, Y_train)
```

## Predicting the Test set results

```python
Y_pred_DTC = classifier_DTC.predict(X_test)
print(Y_pred_DTC)
```

## Making the Confusion Matrix and Accuracy

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test, Y_pred_DTC)
accuracy = accuracy_score(Y_test, Y_pred_DTC)
print(f"Confusion Matrix:\n{cm} \nAccuracy : {accuracy}")
```

# K-Nearest Neighbors Classification Model

## Training model on the Training data set

```python
from sklearn.neighbors import KNeighborsClassifier
classifier_KNN = KNeighborsClassifier( n_neighbors = 5, algorithm = 'kd_tree', leaf_size = 25,  metric = 'minkowski', p = 2)
classifier_KNN.fit(X_train, Y_train)
```

## Predicting the Test set results

```python
Y_pred_KNN = classifier_KNN.predict(X_test)
print(Y_pred_KNN)
```

## Making the Confusion Matrix and Accuracy

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test, Y_pred_KNN)
accuracy = accuracy_score(Y_test, Y_pred_KNN)
print(f"Confusion Matrix:\n{cm} \nAccuracy : {accuracy}")
```
