# Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Importing the Dataset

```python
path = "Data_regression.csv"
df = pd.read_csv(path)
df.head
```

```python
x = df.iloc[:, 0:4].values
y = df.iloc[:, 4].values
print(x)
print(y)
```

# Splitting Dataset into Traning set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

# 1. **Simple Linear Regression Model**

## Training the model on the Training data set

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(X_train,Y_train)
```

## Prediciting the Test Set Results

```python
Y_Predict_LR = clf.predict(X_test)
print(Y_Predict_LR)
```

## Calculate R-squared (R^2) Value

```python
from sklearn.metrics import r2_score

r2 = r2_score(Y_test, Y_Predict_LR)

print("R-squared (R^2) value for Linear Regression:", r2)
```

# 2. **Polynomial Linear Regression Model**

## Training the model on the Training data set

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures( degree = 10)
x_poly = poly_reg.fit_transform(X_train)
linear_reg_2 = LinearRegression()
linear_reg_2.fit(x_poly, Y_train)
```

## Prediciting the Test Set Results

```python
poly_reg = PolynomialFeatures( degree = 10)
x_poly_test = poly_reg.fit_transform(X_test)
Y_Predict_PLR = linear_reg_2.predict(x_poly_test)
print(Y_Predict_PLR)
```

## Calculate R-squared (R^2) Value

```python
from sklearn.metrics import r2_score

r2 = r2_score(Y_test, Y_Predict_PLR)

print("R-squared (R^2) value for Polynomial Linear Regression:", r2)
```

# 3. **Decision Tree Regression Model**

## Training the model on the Training data set

```python
from sklearn.tree import DecisionTreeRegressor

regressor_DTR = DecisionTreeRegressor()
regressor_DTR.fit(X_train, Y_train)
```

## Prediciting the Test Set Results

```python
y_pred_DTR = regressor_DTR.predict(X_test)
print(y_pred_DTR)
```

## Calculate R-squared (R^2) Value

```python
from sklearn.metrics import r2_score

r2 = r2_score(Y_test, y_pred_DTR)

print("R-squared (R^2) value for Decision Tree Regressor:", r2)
```
