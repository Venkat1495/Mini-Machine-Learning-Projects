# Importing Libraries

```python
import pandas as pd
import numpy as np
```

# Importing Dataset

* Reading dataset and print the data in console

```python
path = "Data.csv"
df = pd.read_csv(path)
df.head
```

* Understanding the usage of iloc and loc functions

```python
x = df.iloc[:, 0:3].values
y = df.iloc[:, 3].values
print(x)
print(y)
```

# Taking Care of Missing Data

* To Check number of missing values

```python
df.isnull().sum()
```

* Understand how to fill missing data using sklearn and fillna funtions

```python
from sklearn import impute
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:,1:3])
x[:,1:3] = imputer.transform(x[:,1:3])
print(x)
#df[['Age']] = df[['Age']].fillna(value=df['Age'].mean())
#df[['Salary']] = df[['Salary']].fillna(method = 'pad')
```

# Encoding independent and dependent variables

* Added necessary imports for encoding

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

* Able understand the combine use of columnTransformer and One HotEncoder

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
print(x)
```

* Used the Labelencoder for the following

```python
from sklearn.preprocessing import LabelEncoder
y = LabelEncoder().fit_transform(y)
print(y)
```

# Split dataset into training set and test set

* Again used the sklearn to split the train and test datasets

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

* Used the sklearn preprocessing to
