# Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing the dataset

* Reading dataset and print the data in console

```python
path = "50_Startups_multi.csv"
df = pd.read_csv(path)
df.head
```

* Understanding the usage of iloc and loc functions

```python
x = df.iloc[:, 0:4].values
y = df.iloc[:, 4].values
print(x)
print(y)
```

# Encoding the categorical data

* Using Column Transfomer to work on the state column to change from categorical data to 1, 0 encoding using OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
print(x)
```

# Splitting the dataset into the Training set and Test set

* Using the 'train_test_split' split the data into train data and test data in which random_state helps to get the same data consistently.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

# Training the Multiple Linear Regression model on the Training set

* Using the LinearRegression model to train the training datasets.

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(x_train,y_train)
```

# Predicting the Test set results

* using the 'predict' method to predict results for test dataset and displaying the actual values and prediction values.

```python
Y_Predict = clf.predict(x_test)
Both_Actual_Predict = y_test
BAP = pd.DataFrame(({'Actual': Both_Actual_Predict, 'Prediction': Y_Predict}))
print(BAP)
```
